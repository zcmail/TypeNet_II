```python
def make_DataFrame (pd_data,user_list,train_samples):
    colum_list = ['user','genuine','score','imostor_user']
    df_makeData_all = pd.DataFrame(columns = colum_list)

    for j,u in enumerate(user_list):
        index = 0
        #print(u)
        pd_query = pd_data[pd_data['2'] == u]
        print(pd_query)
        k = len(pd_query)
        index += j*k
        for i in range(k):
            #print(i)
            pd_caclulat = pd_query[pd_query['3'] == str(index+i)]
            #print(pd_caclulat)
            pd_caclulat = pd_caclulat.iloc[:train_samples,4]
            '''
            if u == 's041':
                print(pd_caclulat)
            '''
            dist_ave = pd_caclulat.mean()
            print('u',u,user)
            if u == user:
                df_makeData_all=df_makeData_all.append(pd.DataFrame({'user':[user],'genuine':[0],'score':[dist_ave],'imostor_user':[u]}),ignore_index=True)
            else:
                df_makeData_all=df_makeData_all.append(pd.DataFrame({'user':[user],'genuine':[1],'score':[dist_ave],'imostor_user':[u]}),ignore_index=True)
                
        
    output_fileName = './result_output/TypeNet_1LSTM_3Dense_metric_final_result'+str(user)+'_'+str(train_samples)+'.csv' 
    df_makeData_all.to_csv(output_fileName)
    return df_makeData_all
```
